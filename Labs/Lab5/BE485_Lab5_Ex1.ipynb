## <font color=red>Theory</font>
The sun is actually much hotter on the inside, but we can approximate its surface, *at which the emission takes place*, as a blackbody at a temperature of about <B>5780</B> $^o$K (numbers around <B>6000</B>$^o$K are used sometimes too).<br>
## The Stefan-Boltzmann equation
This equation gives us the energy flux emitted at the sun’s surface:<br>
### <center> M$_s$ = &sigma;.T$^4$<B><font color = blue> [<i>Eq. 1</i>]</font></B></center>
Where:
> &sigma; =  5.6697 × 10$^{-8}$ W·m$^{-2}$·K$^{-4}$, and <br>
> T = 5780$^o$K 

# Inverse Square Distance Law
As this energy propagates through outerspace towards the Earth it weakens as it needs to cover much larger areas and volumes.  It will spread over larger spherical surfaces as it travels outward. The intensity of this radiation decays proportionally to:
## <center> $\frac{1}{r^2}$ <B><font color = blue> [<i>Eq. 2</i>]</font></B></center>
Where :
> r : is the distance away from the source.
<center><img src="https://upload.wikimedia.org/wikipedia/commons/thumb/2/28/Inverse_square_law.svg/420px-Inverse_square_law.svg.png" /></center>

# Cosine Law
The Lambert's cosine law states that the radiant intensity is directly proportional to the cosine of the angle &theta; between the direction of the incident light and the surface normal.
As the angle of incidence changes from 90$^o$ (normal angle) the less effective the transmission. This leads to the simple equation that says:
### <center>E<sub>&theta;</sub> = E. cos(&theta;)<B><font color = blue> [Eq. 3]</font></B></center>
<img src=http://homepages.inf.ed.ac.uk/rbf/CVonline/LOCAL_COPIES/RYER/fig6-3.gif />
With these three Equations, you can estimate the incident irradiance at any point in space at any angle.<br>

```python
# load necessary library modules 
import os
import sys
import numpy as np
import matplotlib.pyplot as plt
import math as math
```

```python
# Now create a figure of the Intensity following the cosine law 

# Here we create a X-Axis from 0 to 90 degrees with 100 steps [array of x]. The linspace does that
x = np.linspace(0, 90, 100)

# conversion betwen degrees and degree Radians for all the whole array [x] created above 
x_Rad=math.pi*x/180

# Cosine function again an array for all the numbers created above [x_Rad]
c = np.cos(x_Rad)
```

```python
#Study and undrstand this code so you can reuse it later

plt.plot(x, c, 'gx') # Plot the c vs x arrays [in degrees] using green color and 'x' markers [gx]. Try [ro]

# Add grids to plot
plt.grid(linestyle=':', linewidth='0.25', color='grey') 

#Add titles to plot
plt.xlabel('Angle', fontsize=12)
plt.ylabel('Intensity', fontsize=12)
plt.title('Cosine Law', fontsize=20)

#plt.tight_layout()
#You can save your figure if you wish 
plt.savefig('My_Cosine_Law_Plot.png')

# The above creates a canvas that need to be plotted now
plt.show()
```

```python
from scipy.constants import codata  # Sci-Python modules contain lots of physical and scientific constants so you do not need to recall them
# Access some physical constants and apply Planks Law 
# https://www.nist.gov/programs-projects/codata-values-fundamental-physical-constants
# https://docs.scipy.org/doc/scipy-0.17.0/reference/constants.html
D = codata.physical_constants  # Where constants are stored

# to inspect what is in this databse of physcial constant use the print command 
#print (D)
```

```python
#Stefan–Boltzmann constant
sigma=D['Stefan-Boltzmann constant'][0]
print ("The Stefan-Boltzmann constant = ",sigma)
```

```python
# T_Sun temperature in Kelvin
T = 5780 # In degree Kelvin

# Sun Emmissivity 
Ms = sigma * math.pow(T,4)

print("The sun's radiation @ Temperatue  = ",T, "K")
#print("Temperature = ",T, "K")
print("Emissive Power M (Watts/m^2)= ", Ms, "or = M (MW/m^2)",Ms/1E6)
```

```python
# Let us now use the databse of astrophysics constants 
# From https://het.as.utexas.edu/HET/Software/Astropy-1.0/constants/index.html
from astropy import constants as const

Rs=const.R_sun
Re=const.R_earth

# Distance Sun to Earth 149597870700 m ~ 149.6E9
Dse=149.6E9

print ("The Sun Radius = ", Rs)
print ("The Earth Radius = ", Re)
print ("Sun Earth Distance = ", Dse)
```

```python
# so at the Earth Surface the irradiance received is
I =  Ms * (Rs*Rs)/(Dse*Dse) 

print("Irradiance at Earth Surface:")
print("Irradiance (W/m^2)=",I,"and in (J/s.m^2)=",I)
#print("Irradiance (J/s.m^2)=",I)
```
