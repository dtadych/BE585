10. ***Eta Carinae*** is a blue hypergiant star, located about 7,500 light-years from the Sun. Its temperature is 66,500 degF, and it is probably one of the hottest stars in the Universe. <br>
Assume:
* σ (the Stefan-Boltzmann constant) = 5.7×10-8 W/m^2xK^4 and
* Wien’s displacement constant, k = 2897 µm. K <br>
<br>
    A. [5] Compute its emissive power (M) and peak emission wavelength (lambda)<br>


```python
import math as math

T = 66500 # degF
sigma = 5.7e-8 # W/m2K4
k = 2897 # micrometers * K

# T_Sun temperature in Kelvin
# Kelvin = (Fahrenheit - 32) / 1.8 + 273.15
T = (T-32)/1.8 + 273.15 # In degree Kelvin

# Sun Emmissivity 
Ms = sigma * math.pow(T,4)

print("The sun's radiation @ Temperature  = ",T, "K")
#print("Temperature = ",T, "K")
print("Emissive Power M (Watts/m^2)= ", Ms, "or = M (MW/m^2)",Ms/1E6)
wav = k/T
print("Maximum wavelength: ", wav, "micrometers")
```

    B. [5] Can you compute how much ‘insolation’ hits the Earth at the Equator (at the spring equinox) at local noon if this was our sun (same size and distance of our actual sun)?

```python
# Let us now use the databse of astrophysics constants 
# From https://het.as.utexas.edu/HET/Software/Astropy-1.0/constants/index.html
from astropy import constants as const
import numpy as np

Rs=const.R_sun
Re=const.R_earth

# Distance Sun to Earth 149597870700 m ~ 149.6E9
Dse=149.6E9

print ("The Sun Radius = ", Rs)
print ("The Earth Radius = ", Re)
print ("Sun Earth Distance = ", Dse)

# so at the Earth Surface the irradiance received is
I =  Ms * (Rs*Rs)/(Dse*Dse) 

print("Irradiance at Earth Surface:")
print("Irradiance (W/m^2)=",I,"and in (J/s.m^2)=",I)
#print("Irradiance (J/s.m^2)=",I)

# Now for Tucson during the spring equinox
lat = 0
tilt = 2
x = lat + tilt
# conversion between degrees and degree Radians for all the whole array [x] created above 
x_rad=math.pi*x/180

print("Irradiance at Equator:")
print("Irradiance (W/m^2)=",I*np.cos(x_rad),"and in (J/s.m^2)=",I*np.cos(x_rad))
#print("Irradiance (J/s.m^2)=",I)
```

```python
# load library modules 
import os
import numpy as np
import matplotlib.pyplot as plt

#Notice this is a new library with additonal functionality 
import viplab_lib2 as vip

# Starts main program
#subset file information
filename="NEON_ExamData.bsq"
nrows=500
ncols=500
nbands=426
datatype=np.int16

# Read all bands into a single 3D DataCube

# Notice the -426  in the call and that means load all bands (all 426 of them)

DataCube=vip.BSQ_band_read(filename,-426,nrows,ncols)

# Assign some of the bands to band/color names that are easier to remember instaead of listing numbers
# Remember these could be from any neighboring bands also since this is hyperspectral 
bandRED=48
bandGREEN=34
bandBLUE=17
bandNIR=97

# The clue here from the exam is a little confusing so I'm going to assume the bands for SWIR1 and SWIR2
# are 5 and 6 (indexing starts at zero though so 4 and 5) because that's what it was for landsat
bandSWIR1 = 4
bandSWIR2 = 5
```

```python
fig, ax = plt.subplots(2, 3,figsize=(12,8)) # a plot with 2 areas labeled 1 and 2 , indexed below 0 and 1

ax[0, 0].imshow(DataCube[:,:,bandRED],cmap='gray');
ax[0, 0].title.set_text('Red Band')

ax[0, 1].imshow(DataCube[:,:,bandGREEN],cmap='gray');
ax[0, 1].title.set_text('Green Band')

ax[0, 2].imshow(DataCube[:,:,bandBLUE],cmap='gray');
ax[0, 2].title.set_text('Blue Band')

ax[1, 0].imshow(DataCube[:,:,bandNIR],cmap='gray');
ax[1, 0].title.set_text('NIR Band')

ax[1, 1].imshow(DataCube[:,:,bandSWIR1],cmap='gray');
ax[1, 1].title.set_text('SWIR1 Band')

ax[1, 2].imshow(DataCube[:,:,bandSWIR2],cmap='gray');
ax[1, 2].title.set_text('SWIR2 Band')
```

14.	[10] Extract, plot, and label the spectral signatures for a Pine Tree, Pecan Tree, Patch of Green Grass, Bare Soil, Dry Grass, Road, Water tank, Roof/House 

```python
# Read the wavelength values from the textfile
file = open(r'NEON_wavelength_values.txt', 'r') 
Xvalues= file.readlines() 
nvalues=len(Xvalues)

#convert text to number (float)
for i in range(0,nvalues):
  Xvalues[i]=float(Xvalues[i]) 

# close text file
file.close

# Extract a single pixel through all bands now to develop a spectral signature
PineTree_Spectra=DataCube[400,400,:]
PecanTree_Spectra=DataCube[400,200,:]
GreenGrass_Spectra=DataCube[220,200,:]
DryGrass_Spectra = DataCube[10,490,:]
Soil_Spectra = DataCube[210,40,:]
Road_Spectra=DataCube[47,16,:]
WaterTank_Spectra=DataCube[200,400,:]

plt.figure(figsize=(16,8))
plt.title('Spectral Signatures',fontsize=20)
plt.xlabel('Wavelength (nm)',fontsize=12)
plt.ylabel('Reflectance ',fontsize=12)
plt.plot(Xvalues,PineTree_Spectra,color='r',label='Pine Tree')
plt.plot(Xvalues,PecanTree_Spectra,color='y',label='Pecan Tree')
plt.plot(Xvalues,GreenGrass_Spectra,color='grey',label='Green Grass')
plt.plot(Xvalues,DryGrass_Spectra,color='m',label='Dry Grass')
plt.plot(Xvalues,Soil_Spectra,color='c',label='Soil')
plt.plot(Xvalues,Road_Spectra,color='k',label='Road')
plt.plot(Xvalues,WaterTank_Spectra,color='b',label='Water Tank')
plt.grid()
plt.legend(loc='upper left')
```
